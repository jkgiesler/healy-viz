```python
import pandas as pd
from gapminder import gapminder
from plotnine import *
%matplotlib inline
```

```python
LOG_MINOR_BREAKS = [300,1000,3000,10000,30000,100000]
GGBLUE='#3d68f6'
```

```python
# 3.4
p = ggplot(data=gapminder,mapping=aes(x='gdpPercap',y='lifeExp'))
p + geom_point()
```

```python
# 3.6
# TODO(jkg): GAM
p + geom_point() + geom_smooth(method='loess',color=GGBLUE)
```

```python
# 3.7
p + geom_point() + geom_smooth(method='lm',color=GGBLUE)
```

```python
def dollar(x): return ['${:,.0f}'.format(y) for y in x]
```

```python
# 3.9

# TODO(jkg): implement GAM as custom smoother method
# geom smooth accepts callable smoother

p + geom_point() + \
    geom_smooth(method='loess',color=GGBLUE) + \
    scale_x_log10(labels = dollar,minor_breaks=LOG_MINOR_BREAKS)
```

```python
# 3.11
p + geom_point(color='#8f43e7') + \
     geom_smooth(method='loess',color=GGBLUE) + \
     scale_x_log10(minor_breaks = LOG_MINOR_BREAKS)
```

```python
# 3.12

p + geom_point(alpha = 0.3) + \
    geom_smooth(
        color='orange', se=False,
        size=8, method='lm'
     ) + \
     scale_x_log10(
         minor_breaks = LOG_MINOR_BREAKS
     )
```

```python
# 3.13

# TODO(jkg): subtitle, caption, GAM
# subtitle/caption: https://github.com/has2k1/plotnine/issues/177
# geom smooth accepts callable smoother


p + geom_point(alpha=0.3) + \
    geom_smooth(method="loess",color=GGBLUE) + \
    scale_x_log10(
        labels=dollar,
        minor_breaks=LOG_MINOR_BREAKS
    ) + \
    labs(
      x = "GDP Per Capita",
      y = "Life Expectancy in Years",
      title = "Economic Growth and Life Expectancy",
      subtitle = "Data points are country-years",
      caption = "Source: Gapminder"
    )
```

```python
# 3.16

GGRED   = '#e87e72'
GGGREENISH = '#a4a433' 
GGGREEN = '#55bb82'
GGTEAL = '#4caef0'
GGLIGHTPURPLE = '#d775ec'


p2  = ggplot(
        data=gapminder,
        mapping=aes(
            x='gdpPercap',
            y='lifeExp',
            color='continent',
            fill='continent'
        )
)
p2 + geom_point() + \
     geom_smooth(method='loess') + \
     scale_x_log10(minor_breaks = LOG_MINOR_BREAKS)+scale_color_manual(
        [GGRED,GGGREENISH,GGGREEN,GGTEAL,GGLIGHTPURPLE]
    )+scale_fill_manual(
        [GGRED,GGGREENISH,GGGREEN,GGTEAL,GGLIGHTPURPLE]
    )
```
